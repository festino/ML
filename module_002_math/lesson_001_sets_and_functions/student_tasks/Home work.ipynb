# Задание 1:
Выполнить операции: $A \cup B; A \cap B; A \setminus B; B \setminus A; A \times B; B \times A$, где A и B

* $A=\{a,1,2\};B=\{a,b,1\}$
* $A=\{2n-1 \mid n \in N\}; B=\{-1,0,1,2,3\}$
* $A=(-\infty , 3); B=[-1, +\infty)$

```python
A = {'a', '1', '2'}
B = {'a', 'b', '1'}
print(A.union(B))
print(A.intersection(B))
print(A.difference(B))
print(B.difference(A))
dec_mult = set()
for a in A:
    for b in B:
        dec_mult.add((a, b))
print(dec_mult)
dec_mult = set()
for b in B:
    for a in A:
        dec_mult.add((b, a))
print(dec_mult)
```

# Задание 2:
Построить графики всех элементарных функций с данного сайта: http://mathprofi.ru/grafiki_i_svoistva_funkcij.html

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
from math import pi

def config_plt(p):
    p.xlim((-5, 5))
    p.ylim((-10, 10))
    #p.axes().set_aspect('equal')
    #p.axes().set_anchor((0, 0))
    p.xlabel('x')
    p.ylabel('y')
    p.grid(True, alpha=0.4)


def f_linear(x):
    return 2 * x - 3

def f_parabola(x):
    return 0.7 * x * x + 2

def f_q_parabola(x):
    return 0.3 * x * x * x - x * x + 1

def f_taylor(x):
    return ((((x/120 + 1/384) * x - 1/6) * x - 1/8) * x + 1) * x + 1

def f_sqrt(x):
    return x ** (1/2)

def f_hyperbola(x):
    return 1 + 2 / (x - 1)

def f_exp(x):
    return math.e ** x

def f_ln(x):
    return np.log(x)


def f_sin(x):
    return np.sin(x)

def f_cos(x):
    return np.cos(x)

def f_tg(x):
    return np.tan(x)

def f_ctg(x):
    return -f_tg(x - pi/2)


def f_arcsin(x):
    return np.arcsin(x)

def f_arccos(x):
    return np.arccos(x)

def f_arctg(x):
    return np.arctan(x)

def f_arcctg(x):
    return f_arctg(-x) + pi/2


x = np.arange(-5, 5.1, 0.2)

# x, x^2, x^3, x^n, sqrt, 1/x, e^x, ln x, trigonometric, arc
fig, (ax1, ax2) = plt.subplots(1, 2)
y = f_linear(x)
#config_plt(plt)
ax1.plot(x, y)
plt.title('Linear')

y = f_parabola(x)
#config_plt(plt)
plt.ylim((-5, 15))
ax2.plot(x, y)
plt.title('Parabola')
plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2)
y = f_q_parabola(x)
#config_plt(plt)
plt.ylim((-15, 15))
ax1.plot(x, y)
plt.title('Qubic parabola')

y = f_taylor(x)
#config_plt(plt)
plt.ylim((-3, 3))
ax2.plot(x, y)
plt.title('Taylor series: sin(x) + cos(x/2)')
plt.show()

x = np.arange(0, 5.1, 0.2)

fig, (ax1, ax2) = plt.subplots(1, 2)
y = f_sqrt(x)
#config_plt(plt)
ax1.plot(x, y)
plt.title('Square root')

x = np.arange(1.2, 5.1, 0.2)
y = f_hyperbola(x)
ax2.plot(x, y)
x = np.arange(-5, 1, 0.2)
y = f_hyperbola(x)
ax2.plot(x, y)
plt.title('Hyperbola')
#config_plt(plt)
plt.show()

x = np.arange(-pi, pi+0.0001, 0.2)

fig, (ax1, ax2) = plt.subplots(1, 2)
y = f_exp(x)
#config_plt(plt)
ax1.plot(x, y)
plt.title('exp')

x = np.arange(0.1, 5.1, 0.2)
y = f_ln(x)
#config_plt(plt)
ax2.plot(x, y)
plt.title('ln')
plt.show()


x = np.arange(-2*pi, 2*pi, 0.1)

fig, (ax1, ax2) = plt.subplots(1, 2)
y = f_sin(x)
ax1.plot(x, y)
ax1.set_title('sin')

y = f_cos(x)
ax2.plot(x, y)
ax2.set_title('cos')
plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2)
x_s = -2.5*pi
while x_s < 2*pi:
    x = np.arange(x_s+0.07, x_s+pi-0.03, 0.1)
    y = f_tg(x)
    ax1.plot(x, y)
    x_s += pi
ax1.set_title('tg')

x_s = -2*pi
while x_s < 2*pi:
    x = np.arange(x_s+0.07, x_s+pi-0.03, 0.1)
    y = f_ctg(x)
    ax2.plot(x, y)
    x_s += pi
ax2.set_title('ctg')
plt.show()


x = np.arange(-1, 1.01, 0.1)

fig, (ax1, ax2) = plt.subplots(1, 2)
y = f_arcsin(x)
ax1.plot(x, y)
ax1.set_title('arcsin')

y = f_arccos(x)
ax2.plot(x, y)
ax2.set_title('arccos')
plt.show()

x = np.arange(-20, 20.01, 0.1)

fig, (ax1, ax2) = plt.subplots(1, 2)
y = f_arctg(x)
ax1.plot(x, y)
ax1.set_title('arctg')

y = f_arcctg(x)
ax2.plot(x, y)
ax2.set_title('arcctg')
plt.show()
```
